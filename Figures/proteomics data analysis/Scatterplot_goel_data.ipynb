```python
preprocessed_data['Mass'] = preprocessed_data['Mass'].str.replace(',','.')
preprocessed_data['Mass']
```

```python
# Convert mass to float. Take average if two masses noted, because of two locus tags
for i in preprocessed_data[pd.to_numeric(preprocessed_data['Mass'], errors='coerce').isnull()==True]["Mass"].index:
    print i
    print preprocessed_data["Mass"][i].split(";")
    a = [float(j) for j in preprocessed_data["Mass"][i].split(";")]
    preprocessed_data["Mass"][i] = sum(a)/len(a)
    
preprocessed_data["Mass"] = pd.to_numeric(preprocessed_data['Mass'], errors='coerce')
```

#### iBAQ based
Protein fraction = iBAQ/sum over all iBAQs

Assumption: all samples have same protein concentration, this is important because of the detection limit of the MS/MS.

```python
# Create a list of indexes for all iBAQ intensity columns
iBAQ_columns = [column for column in preprocessed_data.columns if 'iBAQ ' in column]
iBAQ_columns = [column for column in iBAQ_columns if 'wt' in column]
# Create a new dataframe with only the iBAQ intensity columns
data_iBAQ = preprocessed_data.loc[:, iBAQ_columns]
data_iBAQ = data_iBAQ.apply(pd.to_numeric)
data_iBAQ.head()
```

```python
# change index to locus tags
data_iBAQ_llmg = data_iBAQ.copy()
data_iBAQ_llmg["locus_tag"] = preprocessed_data["Gene names  (ordered locus )"]
data_iBAQ_llmg.set_index("locus_tag", inplace=True)
data_iBAQ_llmg.head()
```

```python
# iBAQ/sum of iBAQs, llmg index
iBAQ_frac_llmg = data_iBAQ_llmg.divide(data_iBAQ_llmg.sum(axis=0), axis=1)
iBAQ_frac_llmg.head()
```

```python
# average fraction
average_frac = iBAQ_frac_llmg.mean(axis=1)
average_frac.head()
```

```python
# order based on decreasing values and get locus tags
print average_frac.sort_values(ascending=False).iloc[0:50].sum()
print average_frac.sort_values(ascending=False).iloc[0:100].sum()
print average_frac.sort_values(ascending=False).iloc[0:109].sum()
```

```python
# order based on decreasing values and get locus tags
largest_fractions = average_frac.sort_values(ascending=False).iloc[0:50].index
largest_fractions = average_frac.sort_values(ascending=False).iloc[0:100].index
largest_fractions = average_frac.sort_values(ascending=False).iloc[0:109].index

largest_fractions
```

```python
# Import Goel data supplementary table 2
goel_data = pd.read_excel(os.path.join(workbookDir,"data", "Goel2015", "mmi_13012_Table_S2.xlsx"))
                          #                          na_filter = False)
goel_data.head()
```

```python
# Bonferroni correction of p-values
goel_data['0.3_adjusted_p'] = goel_data['0.3_pval'].mul(3*len(goel_data))
goel_data['0.5_adjusted_p'] = goel_data['0.5_pval'].mul(3*len(goel_data))
goel_data['0.6_adjusted_p'] = goel_data['0.6_pval'].mul(3*len(goel_data))

goel_data
```

### Data selection based on largest_fraction goel data

```python
# select largest_fractions genes from goel_data
selected_data = goel_data[goel_data["gene"].isin(largest_fractions)]
selected_data.head()
```

```python
# filter selected data on fraction (sol or mem) with highest spectrum count per gene.
selected_genes = list(set(selected_data['gene'].values))
for gene in selected_genes:
    #print selected_data[selected_data['gene']==gene]
    #print selected_data[selected_data['gene']==gene]['spectra'].idxmax()
    #print selected_data[selected_data['gene']==gene]['spectra'].idxmin()
    if not np.isnan(selected_data[selected_data['gene']==gene]['spectra'].idxmin()):
        selected_data.drop(selected_data[selected_data['gene']==gene]['spectra'].idxmin(), inplace=True)
    
selected_data.head()
```

### Plot selections

```python
# Select foldchange columns
sel_dat_transp = selected_data[['0.3_effect', '0.5_effect', '0.6_effect']].transpose()
sel_dat_transp['growth_rate'] = [0.3,0.5,0.6]
sel_dat_transp
```

```python
# Select p-values
sel_dat_transp_p = selected_data[['0.3_pval', '0.5_pval', '0.6_pval']].transpose()
sel_dat_transp_p
```

```python
# Select adjusted p-values
sel_dat_transp_adj_p = selected_data[['0.3_adjusted_p', '0.5_adjusted_p', '0.6_adjusted_p']].transpose()
sel_dat_transp_adj_p
```

```python
# sizing data dots according to t-test significance level
significance_df = pd.DataFrame(np.nan, index=sel_dat_transp_p.index, columns=sel_dat_transp_p.columns)
significance_df

for col in [int(i) for i in selected_data.index]:
    for row in significance_df.index:
        
        if sel_dat_transp_p[col][row] >= 0.05:
            significance_df[col][row] = 'dimgrey' #10
        elif sel_dat_transp_p[col][row] < 0.05:
            significance_df[col][row] = '#d62728' #'tab:green'#30
        #elif sel_dat_transp_p[col][row] < 0.01:
        #    significance_df[col][row] = 'tab:green' #50

significance_df
```

```python
# sizing data dots according to t-test significance level adjusted p-value
significance_df_adj_p = pd.DataFrame(np.nan, index=sel_dat_transp_adj_p.index, columns=sel_dat_transp_adj_p.columns)
significance_df_adj_p

for col in [int(i) for i in selected_data.index]:
    for row in significance_df_adj_p.index:
        
        if sel_dat_transp_adj_p[col][row] >= 0.05:
            significance_df_adj_p[col][row] = 'dimgrey' #10
        elif sel_dat_transp_adj_p[col][row] < 0.05:
            significance_df_adj_p[col][row] = '#d62728' #'#2ca02c' #'#d62728' #'tab:green'#30
        #elif sel_dat_transp_adj_p[col][row] < 0.01:
        #    significance_df_adj_p[col][row] = '#d62728' #'tab:green' #50

significance_df_adj_p
```

```python
# Scatterplot top 100 largest protein fractions; coloured accroding to normal p-value
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Growth rate relative to 0.15 /h', fontsize = 15)
ax.set_ylabel('Log2 foldchange', fontsize = 15)
for col in sel_dat_transp.columns[:-1]:
    #print col
    for rate in sel_dat_transp.index:
        significance_index = rate[:-6]+'pval'
        ax.scatter(sel_dat_transp['growth_rate'][rate],
               sel_dat_transp[col][rate],
               #c='dimgrey',
                   #s=significance_df[col][significance_index]
                s=15,
               c=significance_df[col][significance_index]
              )
    ax.plot(sel_dat_transp['growth_rate'],
               sel_dat_transp[col],
            c='dimgrey',
            alpha=0.5
              )
               #, c = color
ax.axhline(y=1, color='#d62728', ls='--')
ax.axhline(y=-1, color='#d62728', ls='--')

ax.grid()
fig.show()
```

```python
# Scatterplot top 100 largest protein fractions; coloured accroding to adjusted p-value
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Growth rate relative to 0.15 /h', fontsize = 15)
ax.set_ylabel('Log2 foldchange', fontsize = 15)
for col in sel_dat_transp.columns[:-1]:
    for rate in sel_dat_transp.index:
        significance_index = rate[:-6]+'adjusted_p'
        ax.scatter(sel_dat_transp['growth_rate'][rate],
               sel_dat_transp[col][rate],
               #c='dimgrey',
                   #s=significance_df[col][significance_index]
                s=15,
               c=significance_df_adj_p[col][significance_index]
              )
    ax.plot(sel_dat_transp['growth_rate'],
               sel_dat_transp[col],
            c='dimgrey',
            alpha=0.5
              )
ax.axhline(y=1, color='#d62728', ls='--') 
ax.axhline(y=-1, color='#d62728', ls='--')

ax.grid()
fig.show()
```

```python
special_prot_transp_adj_p = special_prot[['0.3_adjusted_p', '0.5_adjusted_p', '0.6_adjusted_p']].transpose()
special_prot_transp_adj_p
```

```python
# sizing data dots according to t-test significance level
significance_df_adj_p_special = pd.DataFrame(np.nan, index=special_prot_transp_adj_p.index, columns=special_prot_transp_adj_p.columns)
significance_df_adj_p_special
significance_df_adj_p_special_size = pd.DataFrame(np.nan, index=special_prot_transp_adj_p.index, columns=special_prot_transp_adj_p.columns)

for col in [int(i) for i in special_prot.index]:
    for row in significance_df_adj_p_special.index:
        if special_prot_transp_adj_p[col][row] > 0.05:
            significance_df_adj_p_special[col][row] = 'dimgrey' #10
            significance_df_adj_p_special_size[col][row] = 10
        #elif special_prot_transp_adj_p[col][row] < 0.05:
        #    significance_df_adj_p_special[col][row] = 'tab:green'#30
        elif special_prot_transp_adj_p[col][row] < 0.05:
            significance_df_adj_p_special[col][row] = 'tab:green' #50
            significance_df_adj_p_special_size[col][row] = 30

significance_df_adj_p_special
significance_df_adj_p_special_size
```

```python
# Scatterplot selected candidates; large dotsize indicates adjusted p-value < 0.05
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Growth rate relative to 0.15 /h', fontsize = 15)
ax.set_ylabel('Log2 foldchange', fontsize = 15)
colors = ['#1f77b4',
          '#ff7f0e',
          '#2ca02c',
          '#d62728',
          '#9467bd',
          '#8c564b',
          '#e377c2',
          '#7f7f7f',
          '#bcbd22',
          '#17becf',
          '#1a55FF']
color_counter=0
for col in special_prot_transp.columns[:-1]:
    color = colors[color_counter]
    color_counter+=1
    for rate in special_prot_transp.index:
        significance_index = rate[:-6]+'adjusted_p'
        ax.scatter(special_prot_transp['growth_rate'][rate],
               special_prot_transp[col][rate],
               c=color,
                s=significance_df_adj_p_special_size[col][significance_index]#,
               #c=significance_df_adj_p_special[col][significance_index]
                   #c='dimgrey'
              )
    ax.plot(special_prot_transp['growth_rate'],
            special_prot_transp[col],
            c=color,
            #label=special_prot.loc[col,'gene']
            label=special_prot.loc[col,'protname']
              )

leg = ax.legend()
ax.grid()
fig.show()
```
